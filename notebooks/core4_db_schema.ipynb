```python
from sqlalchemy import create_engine, text
import pandas as pd

MYSQL_URI = (
    "mysql+pymysql://health_user:strong_password_here"
    "@localhost:3306/HEALTH"
)

engine = create_engine(MYSQL_URI)
```

```python
with engine.connect() as conn:
    result = conn.execute(text("SELECT DATABASE()"))
    print("connected_db:", result.fetchone()[0])
```

테이블 목록 확인

```python
with engine.connect() as conn:
    result = conn.execute(text("SHOW TABLES"))
    tables = [row[0] for row in result.fetchall()]
    print(tables)
```

health_state_log 구조 확인

```python
with engine.connect() as conn:
    result = conn.execute(text("DESCRIBE health_state_log"))
    df_desc = pd.DataFrame(result.fetchall(), columns=result.keys())
    df_desc
```
